# exploratory data analysis (EDA)

```python
import pandas as pd
```

```python
df = pd.read_csv('/home/carolina/lgcm/dev/dvc-in-action/data/data.csv')
```

```python
print("Dimensões do dataset:", df.shape)
print("\nVisualizar as primeiras linhas:")
df.head()
```

```python
print("\nProporção de valores nulos por coluna:")
print((df.isnull().sum() / len(df)) * 100)
```

```python
limite_nulos = 0.6
colunas_para_remover = df.columns[df.isnull().mean() > limite_nulos]

# Exibir as colunas que serão removidas
print("\nColunas com mais de 60% de nulos:")
print(colunas_para_remover)

# Remover as colunas com muitos nulos
df = df.drop(columns=colunas_para_remover)
```

```python
# colunas menos importantes (baseado em entendimento do dataset)
colunas_irrelevantes = ['Id', 'ContentLicense', 'OwnerDisplayName','OwnerUserId','LastEditorUserId']
colunas_para_remover = [col for col in colunas_irrelevantes if col in df.columns]

print("\nColunas irrelevantes:")
print(colunas_para_remover)

# Remover colunas irrelevantes
df = df.drop(columns=colunas_para_remover)
```

```python
df.head()
```

```python
#tratar nulos da coluna de respostas aceitas e transformar em zeros 
df['AcceptedAnswerId'] = df['AcceptedAnswerId'].fillna(0)
```

```python
# Converter colunas de data para datetime
datas = ['CreationDate', 'LastEditDate', 'LastActivityDate']
for coluna in datas:
    df[coluna] = pd.to_datetime(df[coluna])

## Criar novas features baseadas nas datas
df['CreationYear'] = df['CreationDate'].dt.year
df['CreationMonth'] = df['CreationDate'].dt.month
df['CreationDay'] = df['CreationDate'].dt.day
```

```python
# Remover os '<>' e transformar em listas
df['Tags'] = df['Tags'].str.replace('<', '').str.split('>')[:-1]
```
